```python
training_pairs = [data.example_to_training_pair(example, 
                      data.get_word_to_idx(embedding),
                      data.get_entity_to_idx(all_entities))
                  for example in train_data['GetWeather']                
                 ]
training_pairs = [tp for tp in training_pairs if tp != []]
```

```python
training_pairs
```

```python
training_pairs = [data.transpose_training_pair(tp) for tp in training_pairs]
```

```python
import multiplai.evals.nlu.model as model
```

```python
n_words = len(embedding.token_to_idx.keys())
n_labels = len(all_entities)
```

```python
#ctx = mx.gpu(0)
ctx = mx.cpu()

# encoder = EncoderRNN(embedding, hidden_size=32)
# decoder = Decoder(n_labels, context_size=32)

classifier = model.ClassifierRNN(embedding, hidden_size=32, max_labels=n_labels)


model.train_iters(training_pairs, classifier, ctx)
```

```python
%debug
```

```python
model.eval(training_pairs[0:], classifier, ctx, debug=False)
```
